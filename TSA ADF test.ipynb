```python
def adf_test(dataset,column):
    from statsmodels.tsa.stattools import adfuller as adf
    result=adf(dataset[column])
    output = {'ADF Statistic': result[0],'p-value': result[1],'Lags Used': result[2],'Observations': result[3],'Critical Values': result[4]}
    if output['p-value'] <= 0.05:
        print("p-value:",output[p-value])
        print("Data is likely stationary (reject null hypothesis)")
    else:
        print("p-value:",output['p-value'])
        print("Data is non-stationary (fail to reject null hypothesis)")

    import matplotlib.pyplot as plt
    plt.plot(dataset.index,dataset[column],label=column)
    plt.legend(loc='best')
    plt.savefig("{}_ADF.png".format(column))
    plt.show()
```

```python
adf_test(dataset,"Close")
```

```python
import numpy as np
def ApEn(series,m,r):
    import numpy as np
    series=np.array(series)
    
    def phi(x):
        n=len(x)
        c=np.zeros(n)
        for i in range(n):
            d=np.max(np.abs(x-x[i]),axis=1)<=r
            c[i]=np.sum(d)/n
        return np.mean(-np.log(c))

    N=len(series)
    xm=np.array([series[i:i+m] for i in range(N-m+1)])
    xm1=np.array([series[i:i+(m+1)] for i in range(N-(m+1)+1)])
    
        
    return abs(phi(xm)-phi(xm1))
```

```python
ApEn(dataset.Close,2,0.2*np.std(dataset.Close))
```

```python
def SampEn(series,m,r):
    series=np.array(series)
    def phi(x):
        n=len(x)
        c=0
        for i in range(n):
            for j in range(i+1,n): 
                d=np.max(np.abs(x[i]-x[j]))
                if d<=r:
                    c+=1
        return c/(n*(n-1)/2)

    N=len(series)
    xm=np.array([series[i:i+m] for i in range(N-m+1)])
    xm1=np.array([series[i:i+(m+1)] for i in range(N-(m+1)+1)])
    B=phi(xm)
    A=phi(xm1)
    if A==0 or B==0:
        return np.inf
    else:
        return -np.log(A/B)
        
```
