```python
def get_pdf(dataset,start_range,end_range):
    mean=dataset.mean()
    std=dataset.std()
    from scipy.stats import norm
    import numpy as np
    dist=norm(mean,std)
    prb=[dist.pdf(v) for v in range(start_range,end_range)]
    probability=np.sum(prb)
    print("Mean:{}\nStandard Deviation: {}".format(mean,std))
    print("Probability:",probability)

    import seaborn as sns
    import matplotlib.pyplot as plt
    sns.distplot(dataset,kde=True,kde_kws={'color':'blue'},color='Green')
    plt.axvline(start_range,color='red')
    plt.axvline(end_range,color='red')
    plt.title("Distribution of {}".format(dataset.name))
    plt.show()
```

```python
get_pdf(dataset['Eraiyur_Qty'],500,1200)
```

```python
def freq_table(dataset,column):
    import pandas as pd
    freq_table={'Unique Values':dataset[column].value_counts().index,'Frequency':dataset[column].value_counts().values,'Relative Frequency':dataset[["Eraiyur_Qty"]].value_counts().values/len(dataset)
           }
    freq_table=pd.DataFrame(freq_table)
    freq_table['Cumulative Frequency']=freq_table["Relative Frequency"].cumsum()
    #freq_table["Frequency"]=dataset[column].value_counts().values()
    return freq_table
```

```python
def sampen(series,m,r):
    import numpy as np
    series=np.array(series.values)
    N=len(series)
    xm=np.array([series[i:i+m] for i in range(N-m+1)])
    xm1=np.array([series[i:i+m+1] for i in range(N-(m+1)+1)])

    def phi(x):
        c=0
        n=len(x)
        for i in range(n):
            for j in range(i+1,n):
                diff=np.max(x[i]-x[j])
                if diff<=r:
                    c+=1
        return c/(n*(n-1)/2)
    A=phi(xm)
    B=phi(xm1)
    if A==0 or B==0:
        return np.inf
    else:
        return -np.log(B/A)
```

```python
def minmax_split(dataset,column):
    from sklearn.preprocessing import MinMaxScaler
    ms=MinMaxScaler()
    data=ms.fit_transform(dataset[[column]])
    size=int(len(data)*0.80)
    train=data[:size]
    test=data[size:]
    return train,test,ms,size
```

```python
train,test,ms,size=minmax_split(dataset,"Eraiyur_TS")
```

```python
train.shape
```

```python
def eval(model,test,ms):
    pred=model.predict(len(train),len(dataset)-1)
    from sklearn.metrics import mean_squared_error as mse,mean_absolute_error as mape
    Mse=mse(test,pred)
    MaPe=mape(test,pred)
    return Mse,MaPe,pred

def plot(test,pred,ms):
    test1=ms.inverse_transform(test)
    yp=ms.inverse_transform([pred])
    import matplotlib.pyplot as plt
    ind=dataset["Date"][size:]
    plt.plot(ind,test1,label="Actual",color='blue')
    plt.plot(ind,yp.T,label='Predicted',color='red')
    plt.title("Actual vs Predicted")
    plt.legend()
    plt.show()
```

```python
#Auto regression
from statsmodels.tsa.arima.model import ARIMA
trend=['t','c','n','ct']
for td in trend:
    m=ARIMA(train,order=(2,0,0),trend=td)
    model=m.fit()
    Mse,MaPe,pred=eval(model,test,ms)
    print("Trend: {}".format(td))
    print("Mse: {}\nMape : {}".format(Mse,MaPe))
    print("___________________________") 

    plot(test,pred,ms)
    
```

```python
#Integration
from statsmodels.tsa.arima.model import ARIMA
trend=['t','n']
for td in trend:
    m=ARIMA(train,order=(0,1,0),trend=td)
    model=m.fit()
    Mse,MaPe,pred=eval(model,test,ms)
    print("Trend: {}".format(td))
    print("Mse: {}\nMape : {}".format(Mse,MaPe))
    print("___________________________") 

    plot(test,pred,ms)
```

```python
#Moving Average
from statsmodels.tsa.arima.model import ARIMA
trend=['t','n']
for td in trend:
    m=ARIMA(train,order=(0,0,2),trend=td)
    model=m.fit()
    Mse,MaPe,pred=eval(model,test,ms)
    print("Trend: {}".format(td))
    print("Mse: {}\nMape : {}".format(Mse,MaPe))
    print("___________________________") 

    plot(test,pred,ms)
```

```python
order=[(0,0,0),(2,0,1),(1,1,1)]
from statsmodels.tsa.statespace.sarimax import SARIMAX
for i in order:
    m=SARIMAX(train,order=i,seasonal_order=(0,0,0,12))
    model=m.fit()
    Mse,MaPe,pred=eval(model,test,ms)
    print("Trend: {}".format(td))
    print("Mse: {}\nMape : {}".format(Mse,MaPe))
    print("___________________________") 

    plot(test,pred,ms)
```
